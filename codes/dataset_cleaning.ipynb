```python
import pandas as pd
import os
import shutil
```

```python
image_dir = '../odir/images'
csv_file = '../odir/full_df.csv'
data_dir = '../dataset'
```

```python
df = pd.read_csv(csv_file)
df.head()
```

```python
for index, column in df.iterrows():
  filename = column['filename']
  label = column['labels'][2]

  if label == 'A':
    label = 'age_macular_degeneration'
  elif label == 'M':
    label = 'pathological_myopia'
  elif label == 'H':
    label = 'hypertension'
  else:
    continue
  
  
  class_dir = os.path.join(data_dir, label)
  if not os.path.exists(class_dir):
    os.makedirs(class_dir)

  source_path = os.path.join(image_dir, filename)
  dest_path = os.path.join(class_dir, filename)
  shutil.copy(source_path, dest_path)
  
  # if index == 5:
  #   break
```
